# **1. Сингулярные базисы и сингулярные числа матрицы**
**1. Основная теорема**

**Теорема 1.** Пусть $A$ — произвольная матрица из $M_{m,n}$. Существуют такие ортонормированные базисы {$e^k $}$^n_{k=1}$ $⊂$ $C^n$, {$q^k $}$^n_{k=1}$ $⊂$ $C^m$ и положительные числа $σ_1 ≥ σ_2 ≥σ_r$, $0 ≤ r ≤$ min{m,n}, что

\begin{align}
  \text{$Ae^k$} = \left\{
  \begin{array}{cl}
  σ_kq^k, & k ≤ r, \\
  0, & k > r,
  \end{array}
  \right.(1)
\end{align}

Числа $σ_1,,...,σ_r$ называют сингулярными числами матрицы $A$. Базисы {$e^k $}$^n_{k=1}$, {$q^k $}$^n_{k=1}$, обеспечивающие выполнение соотношений (1), называют сингулярными базисами матрицы $A$. Понятно, что $r$ есть размерность Im($A$), т. е. $r$ — ранг матрицы $A$.

**Доказательство теоремы 1.** Матрица $A^∗A$ самосопряжена и неотрицательно
определена. Действительно $(A^∗A)^∗ = A^∗A$, $(A^∗Ax,x)= (Ax,Ax) ≥ 0$  для любого $x ∈ C^n$. Поэтому существует ортонормированный базис собственных векторов {$e^k $}$^n_{k=1}$ матрицы $A^∗A$. Все ее собственные числа неотрицательны. Таким образом,

$$A^*Ae^k=σ_k^2e^k, k=1,2,...,n (2)$$    

$σ^2_k ≥ 0$, $k = 1,2,...,n$. Будем считать при этом, что $σ_1 ≥ σ_2 ≥ ··· ≥ σ_r > 0$, $σ_{r+1} = ··· = σ_n = 0$. Положим $z_k = Ae^k$ для $k = 1,2,...,r$ и заметим, что

$$(z^p, z^q) = (Ae^p,Ae^q) = (A∗Ae^p,e^q) = σ^2_p(e^p,e^q).$$

Значит,

\begin{align}
  \text{$(z^p, z^q)$} = \left\{
  \begin{array}{cl}
  0, & p \neq q, \\
  σ^2_p, & p = q,
  \end{array}
  \right.(3)
\end{align}

следовательно, векторы

$$q^k=σ^{-1}_kAe^k, k=1,2,...,r (4)$$

образуют ортонормированную систему в пространстве $C^m$. Если окажется, что $r$
меньше $m$, дополним ее произвольно векторами $q^k, k = r + 1, r + 2,...,m,$ до
ортонормированного базиса пространства $C^m$. Из равенств (4) вытекает справедливость (1).

**2. Матричное представление сингулярного разложения.** Равенства (1) часто записывают в виде

$$A=VΣW^∗.(5)$$

Здесь $V$ — матрица, столбцами которой служат векторы {$q^k$}$^m_{k=1}$; $W$ — матрица, столбцами которой являются векторы {$e^k$}$^n_{k=1}$; $Σ$ — матрица, у которой главный минор порядка $r$ диагональный, на его диагонали расположены сингулярные числа матрицы $A$, все остальные элементы матрицы $Σ$ — нули.

Именно в такой форме обычно реализуют сингулярное разложение в пакетах стандартных программ линейной алгебры. Например, в среде Matlab сингулярное разложение может быть выполнено при помощи функции **svd** (от английского SVD —
Singular Value Decomposition).

**3. Тензорное представление сингулярного разложения.** Из равенств (1) следует, что

\begin{align}
  Ax = \sum_{j=1}^r σ_j(x,e^j)q^j ∀x ∈ C^n.(6)
\end{align}

Введем в рассмотрение матрицы $q_j ⊗ e_j$ порядка $m$ на $n$, определяемые тождествами

$$q^j ⊗ e^jx = (x,e^j)q^j ∀x ∈ Cn.(7)$$

Матрицу $q^j ⊗e^j$ называют тензорным произведением векторов $q^j, e^j$. Из тождеств (6), (7) получаем, что

\begin{align}
  A = \sum_{j=1}^r σ_jq^j ⊗ e^j.(8)
\end{align}

Равенство (8) есть тензорное представление матрицы $A$. Отметим, что для задания матрицы A, достаточно хранить в памяти компьютера Q = r(m+n+1) чисел. Элементарные оценки показывают, что, например, для квадратной матрицы экономия
достигается, если ранг матрицы не превышает половины ее порядка.

**4. Оценки сингулярных чисел.**  Вследствие (2) очевидным образом получаем, что $\begin{align}
      (A^*Ax,x) = \sum_{j=1}^r σ^2_j|(x, e^j)|^2,
    \end{align}$
поэтому

$$σ^2_1 = \max_{|x|=1}(A^∗Ax,x).(9)$$

Если $A$ — квадратная матрица, то из равенства (9) нетрудно вывести следующие
оценки

$$\max_{1≤i,j≤n}|a_{ij}|≤σ_1≤n\max_{1≤i,j≤n}|a_{ij}|(10)$$

Оценки (10) неулучшаемы. Левое неравенство превращается в равенство при $A = I$, где $I$ — единичная матрица, правое — при $A = E$, где $E$ — матрица, все элементы которой равны единице.
# **2. Некоторые применения сингулярного разложения**
**1. Полярные разложения.** Пусть $A$ — произвольная квадратная матрица. Ра
венство (5) можно переписать в виде

$$A=US, A=TU, (11)$$

где $U = VW^∗$ — унитарная матрица, $S = WΣW^∗$, $T = VΣV^∗$ — самосопряженные
нетрицательные матрицы$^3$. Формулы (11) определяют *полярные разложения* матрицы (левое и правое). Они находят разнообразные приложения, например, в механике деформируемого твердого тела.

**2. Решение систем линейных алгебраических уравнений.** Если сингулярное разложение матрицы $A$ построено, то решение системы линейных алгебраических
уравнений

$$Ax = b, (12)$$

где $b$ — заданный вектор из $C^m$, $x ∈ C^n$ — искомый вектор, строится без труда. В самом деле, представим вектор $b$ в виде
$\begin{align}
  b = \sum_{k=1}^m (b, q^k)q^k,
\end{align}$
а вектор $x$ будем разыскивать в виде разложения
$\begin{align}
  x = \sum_{k=1}^n ξ_ke^k,
\end{align}$
Для определения чисел $ξ_k$, используя (12), получим соотношения $ξ_k = (b,q^k)/σ_k, k = 1,2,...,r$. Далее нужно различать два случая:

1) все числа $η_k = (b,q^k), k = r+1,r+2,...,m$, — нули, тогда

\begin{align}
  x = \sum_{k=1}^r (b, q^k)σ_ke^k+\sum_{k=1+1}^n ξ_ke^k,(13)
\end{align}

где числа $ξ_k, k = r+1,r+2,...,n$, произвольны, есть общее решение системы уравнений (12);

2) если хотя бы одно из чисел $η_k$ отлично от нуля, то система (12) не имеет решений, вектор, определяемый формулой (13), называют в этом случае псевдорешением системы (12); легко проверить, что этот вектор дает не большую длину вектору невязки $r = Ax−b$ по сравнению с любым другим вектором пространства $C^n$. Если сингулярное разложение матрицы $A$ построено, то вычисления по формулам (13) требуют дополнительно всего порядка *rm* арифметических операций.

**3. Оценка возмущения решения системы линейных алгебраических уравнений за счет неточности задания правой части.** Пусть $A$ — квадратная невырожденная матрица, $x$ — решение системы уравнений (12), $\tilde{x}$ — решение системы уравнений $Ax = \tilde{b}$. Тогда $A(x−\tilde{x}) = b−\tilde{b}$. Величину $δ_x = |x−\tilde{x}|/|x|$ называют величиной *относительного изменения решения* при изменении правой части. Выясним, как она зависит от $δ_b = |b −\tilde{b}|/|b|$ — величины *относительного изменения правой части*. Представляя $x$, $\tilde{x}$ в виде разложений по базису {$e^k$}$^n_{k+1}$, а $b$, $\tilde{b}$ — в виде
разложений по базису {$q^k$}$^n_{k+1}$, получим, что

\begin{align}
  δ^2_x = \frac{\sum_{k=1}^n \frac{|η_k-\tilde{η}_k|^2}{σ^k_2}}{\sum_{k=1}^n \frac{|η_k|^2}{σ^k_2}} ≤ \frac{σ^2_1}{σ^2_n} \frac{\sum_{k=1}^n |η_k-\tilde{η}_k|^2}{\sum_{k=1}^n |η_k|^2} = \frac {σ^2_1}{σ^2_n}δ^2_b.(14)
\end{align}

Таким образом,

$$δ_x ≤ \frac{σ_1}{σ_n}δ_b.(15)$$

Величину σ_1/σ_n, характеризующую устойчивость решения уравнения (12) по отношению к изменению его правой части, называют *числом обусловленности* матрицы $A$ и обозначают через cond($A$). Очевидно, cond($A$) $≥$ 1 для любой матрицы $A$.

оценка (15) неулучшаема в том смысле, что для любой невырожденной матрицы $A$
можно указать такие $b$ и $\tilde{b}$, что неравенство (15) превращается в равенство.

**4. Малоранговые приближения матриц.** Как было отмечено выше, при использовании сингулярного разложения количество чисел, требуемых для задания
матрицы, пропорционально ее рангу. В то же время, во многих случаях требуется
лишь приближенное задание матрицы. В связи этим естественным образом возникает следующая задача. Пусть $A ∈ M_{m,n}$. Требуется построить матрицу $X ∈ M_{m,n}$, являющуюся наилучшим приближением к $A$ среди всех матриц, ранг которых не превосходит заданного числа $k$. Пусть $A$, матрица ранга r ≥ 2, представлена в виде (8), и $k < r$. Положим

\begin{align}
  X = \sum_{k=1}^n σ_jq^j ⊗ e^j.(16)
\end{align}

Имеем rank($X$) = $k$, причем
$\begin{align}
  A - X = \sum_{k=1}^n σ_jq^j ⊗ e^j,
\end{align}$
следовательно, максимальное сингулярное число матрицы $A - X$ равно $σ_{k+1}$. Пусть теперь матрица $Y ∈ M_{m,n}$ имеет ранг, не превосходящий $k$. Тогда def($Y$) не меньше, чем $n − k$. Поэтому существует вектор $z \neq 0,z ∈$ span{$e^1,e^2,...,e^{k+1}$}$∩$Ker($Y$). Можно считать, что |$z$| = 1. Положим
$\begin{align}
  z = \sum_{j=1}^{k+1} α_je^j.
\end{align}$
Элементарные вычисления дают, что

\begin{align}
   ((A −Y)^∗(A−Y)z,z) = (A^∗Az,z) = \sum_{j=1}^{k+1} σ_j^2α_j^2 ≥ σ_{k+1}^2.
\end{align}

Вследствие (9) отсюда вытекает, что максимальное сингулярное число матрицы $A−Y$
не меньше, чем $σ_{k+1}$. Таким образом, можно сказать, что в определенном здесь смысле матрица $X$ — наилучшее приближение к матрице $A$ на множестве всех матриц ранга, не превосходящего $k < r$. Полученный результат составляет содержание *теоремы Эккарта — Янга* [1].

Говорят, что матрица $X$ осуществляет сжатие информации, содержащейся в матрице $A$. Если $A$ — квадратная матрица, то, используя оценку (10), близость матриц $X$ и $A$ можно характеризовать неравенством
$$\max_{1≤i,j≤n}| a_{ij}-x_{ij}|≤σ_{k+1}.$$

Код на Python для сжатия изображения:

Ниже приведён первый пример, в котором демонстрируется, как с помощью NumPy можно вычислить сингулярное разложение (SVD) простой матрицы, восстановить её по разложению и визуализировать сингулярные числа.

---

### Пример 1. Сингулярное разложение простой матрицы

В этом примере мы рассмотрим матрицу

\[
A = \begin{pmatrix} 3 & 1 \\ 1 & 3 \end{pmatrix}.
\]

Суть метода SVD состоит в том, что любую матрицу \( A \) можно представить в виде

\[
A = U \Sigma V^T,
\]

где:
- \( U \) — унитарная (или ортогональная, если работаем с вещественными числами) матрица, столбцы которой являются левыми сингулярными векторами;
- \( \Sigma \) — диагональная матрица, содержащая неотрицательные сингулярные числа;
- \( V^T \) — транспонированная унитарная матрица, столбцы которой (до транспонирования) являются правыми сингулярными векторами.

После разложения мы можем восстановить исходную матрицу и сравнить результаты, а также построить столбчатую диаграмму сингулярных чисел.

Ниже представлен соответствующий код на Python:

```python
import numpy as np
import matplotlib.pyplot as plt

def example_svd():
    # Определим простую матрицу A (2x2)
    A = np.array([[3, 1],
                  [1, 3]])
    
    # Вычисляем сингулярное разложение матрицы A
    # U - матрица левых сингулярных векторов,
    # s - вектор сингулярных чисел,
    # Vh - транспонированная матрица правых сингулярных векторов.
    U, s, Vh = np.linalg.svd(A)
    
    # Выводим результаты SVD
    print("Исходная матрица A:")
    print(A)
    print("\nМатрица U (левые сингулярные векторы):")
    print(U)
    print("\nСингулярные числа:")
    print(s)
    print("\nМатрица V^T (транспонированные правые сингулярные векторы):")
    print(Vh)
    
    # Для восстановления A создаём диагональную матрицу Sigma из вектора s.
    S = np.zeros((A.shape[0], A.shape[1]))
    np.fill_diagonal(S, s)
    
    # Восстанавливаем матрицу A по формуле A = U * Sigma * V^T
    A_reconstructed = U @ S @ Vh
    print("\nВосстановленная матрица A:")
    print(A_reconstructed)
    
    # Визуализируем сингулярные числа
    plt.figure(figsize=(6,4))
    plt.bar(range(1, len(s)+1), s, color='skyblue')
    plt.xlabel('Индекс сингулярного числа')
    plt.ylabel('Значение сингулярного числа')
    plt.title('Сингулярные числа матрицы A')
    plt.xticks(range(1, len(s)+1))
    plt.show()

# Запускаем пример
example_svd()
```

---

### Ключевые моменты данного примера

1. **Вычисление SVD:**  
   Функция `np.linalg.svd` возвращает три компонента: \( U \), вектор сингулярных чисел \( s \) и \( V^T \). Вектор \( s \) затем преобразуется в диагональную матрицу \( \Sigma \).

2. **Восстановление матрицы:**  
   Произведением \( U \Sigma V^T \) получается восстановленная матрица, которая должна совпадать с исходной \( A \) (с учётом численных погрешностей).

3. **Визуализация:**  
   С помощью библиотеки Matplotlib строится столбчатая диаграмма, позволяющая увидеть величины сингулярных чисел. Это помогает понять, какая доля информации содержится в каждом компоненте разложения.

Этот пример является отправной точкой для понимания работы SVD и может служить основой для дальнейших экспериментов, таких как создание низкоранговых аппроксимаций или применение SVD в других задачах. Если всё понятно, можем переходить к следующему примеру.

````python
import numpy as np
import matplotlib.pyplot as plt

def example_svd():
    # Определим простую матрицу A (2x2)
    A = np.array([[3, 1],
                  [1, 3]])

    # Вычисляем сингулярное разложение матрицы A
    # U - матрица левых сингулярных векторов,
    # s - вектор сингулярных чисел,
    # Vh - транспонированная матрица правых сингулярных векторов.
    U, s, Vh = np.linalg.svd(A)

    # Выводим результаты SVD
    print("Исходная матрица A:")
    print(A)
    print("\nМатрица U (левые сингулярные векторы):")
    print(U)
    print("\nСингулярные числа:")
    print(s)
    print("\nМатрица V^T (транспонированные правые сингулярные векторы):")
    print(Vh)

    # Для восстановления A создаём диагональную матрицу Sigma из вектора s.
    S = np.zeros((A.shape[0], A.shape[1]))
    np.fill_diagonal(S, s)

    # Восстанавливаем матрицу A по формуле A = U * Sigma * V^T
    A_reconstructed = U @ S @ Vh
    print("\nВосстановленная матрица A:")
    print(A_reconstructed)

    # Визуализируем сингулярные числа
    plt.figure(figsize=(6,4))
    plt.bar(range(1, len(s)+1), s, color='skyblue')
    plt.xlabel('Индекс сингулярного числа')
    plt.ylabel('Значение сингулярного числа')
    plt.title('Сингулярные числа матрицы A')
    plt.xticks(range(1, len(s)+1))
    plt.show()

# Запускаем пример
example_svd()
````

````python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def denoise_image_with_svd_soft(image_path, tau, use_color=False):
    """
    Удаляет шум из изображения с помощью SVD с мягким пороговым значением (soft thresholding).

    Параметры:
    - image_path: путь к изображению.
    - tau: пороговое значение, которое вычитается из каждого сингулярного числа.
    - use_color: если True, обрабатывает изображение по каналам (цветное); если False – в оттенках серого.
    """

    if use_color:
        img = cv2.imread(image_path)
        if img is None:
            print(f"Ошибка: не удалось прочитать изображение {image_path}")
            return
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        channels = cv2.split(img)
        denoised_channels = []
        for idx, channel in enumerate(channels):
            channel_float = np.float32(channel)
            U, s, Vt = np.linalg.svd(channel_float, full_matrices=False)
            # Применяем soft thresholding ко всем сингулярным числам
            s_thresholded = np.maximum(s - tau, 0)
            S_new = np.diag(s_thresholded)
            channel_denoised = U @ S_new @ Vt
            channel_denoised = np.clip(channel_denoised, 0, 255).astype(np.uint8)
            denoised_channels.append(channel_denoised)
        img_denoised = cv2.merge(denoised_channels)
    else:
        img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
        if img is None:
            print(f"Ошибка: не удалось прочитать изображение {image_path}")
            return
        img_float = np.float32(img)
        U, s, Vt = np.linalg.svd(img_float, full_matrices=False)
        s_thresholded = np.maximum(s - tau, 0)
        S_new = np.diag(s_thresholded)
        img_denoised = U @ S_new @ Vt
        img_denoised = np.clip(img_denoised, 0, 255).astype(np.uint8)

    # Визуализация
    plt.figure(figsize=(12, 6))
    if use_color:
        plt.subplot(1, 2, 1)
        plt.imshow(img)
        plt.title("Исходное изображение")
        plt.axis('off')
        plt.subplot(1, 2, 2)
        plt.imshow(img_denoised)
        plt.title(f"Денойзинг (tau={tau})")
        plt.axis('off')
    else:
        plt.subplot(1, 2, 1)
        plt.imshow(img, cmap='gray')
        plt.title("Исходное изображение")
        plt.axis('off')
        plt.subplot(1, 2, 2)
        plt.imshow(img_denoised, cmap='gray')
        plt.title(f"Денойзинг (tau={tau})")
        plt.axis('off')

    plt.show()
    return img_denoised

_img = '/content/photo_2025-02-09_11-33-20.jpg'

# Пример использования:
# Для изображения с небольшими шумами можно попробовать tau = 10, 20 и т.д.
denoise_image_with_svd_soft(_img, tau=30, use_color=True)
````

````python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def denoise_image_with_svd_soft(image_path, tau, use_color=False):
    """
    Удаляет шум из изображения с помощью SVD с мягким пороговым значением (soft thresholding).

    Параметры:
    - image_path: путь к изображению.
    - tau: пороговое значение, которое вычитается из каждого сингулярного числа.
    - use_color: если True, обрабатывает изображение по каналам (цветное); если False – в оттенках серого.
    """

    if use_color:
        # Загружаем цветное изображение (OpenCV читает в формате BGR)
        img = cv2.imread(image_path)
        if img is None:
            print(f"Ошибка: не удалось прочитать изображение {image_path}")
            return
        # Преобразуем BGR в RGB для корректного отображения в matplotlib
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        # Разбиваем изображение на каналы
        channels = cv2.split(img)
        denoised_channels = []
        singular_values_list = []  # для визуализации сингулярных чисел по каждому каналу
        for idx, channel in enumerate(channels):
            channel_float = np.float32(channel)
            U, s, Vt = np.linalg.svd(channel_float, full_matrices=False)
            singular_values_list.append(s)
            # Применяем soft thresholding: вычитаем tau, но не допускаем отрицательных значений
            s_thresholded = np.maximum(s - tau, 0)
            S_new = np.diag(s_thresholded)
            channel_denoised = U @ S_new @ Vt
            channel_denoised = np.clip(channel_denoised, 0, 255).astype(np.uint8)
            denoised_channels.append(channel_denoised)
        # Объединяем обработанные каналы обратно
        img_denoised = cv2.merge(denoised_channels)

        # Визуализируем сингулярные числа для первого канала в качестве примера
        plt.figure(figsize=(10, 4))
        plt.plot(singular_values_list[0], 'bo-', label='Исходные s')
        plt.plot(np.maximum(singular_values_list[0] - tau, 0), 'ro-', label='Пороговые s')
        plt.xlabel('Индекс')
        plt.ylabel('Сингулярное число')
        plt.title('Сингулярные числа (пример для первого канала)')
        plt.legend()
        plt.grid(True)
        plt.show()
    else:
        # Загружаем изображение в оттенках серого
        img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
        if img is None:
            print(f"Ошибка: не удалось прочитать изображение {image_path}")
            return
        img_float = np.float32(img)
        U, s, Vt = np.linalg.svd(img_float, full_matrices=False)

        # Строим график сингулярных чисел для исходного изображения и после soft thresholding
        plt.figure(figsize=(10, 4))
        plt.plot(s, 'bo-', label='Исходные s')
        plt.plot(np.maximum(s - tau, 0), 'ro-', label='Пороговые s')
        plt.xlabel('Индекс')
        plt.ylabel('Сингулярное число')
        plt.title('Сингулярные числа изображения')
        plt.legend()
        plt.grid(True)
        plt.show()

        # Применяем soft thresholding
        s_thresholded = np.maximum(s - tau, 0)
        S_new = np.diag(s_thresholded)
        img_denoised = U @ S_new @ Vt
        img_denoised = np.clip(img_denoised, 0, 255).astype(np.uint8)

    # Отображаем исходное и обработанное изображения
    plt.figure(figsize=(12, 6))
    if use_color:
        plt.subplot(1, 2, 1)
        plt.imshow(img)
        plt.title("Исходное изображение")
        plt.axis('off')
        plt.subplot(1, 2, 2)
        plt.imshow(img_denoised)
        plt.title(f"Денойзинг (tau = {tau})")
        plt.axis('off')
    else:
        plt.subplot(1, 2, 1)
        plt.imshow(img, cmap='gray')
        plt.title("Исходное изображение")
        plt.axis('off')
        plt.subplot(1, 2, 2)
        plt.imshow(img_denoised, cmap='gray')
        plt.title(f"Денойзинг (tau = {tau})")
        plt.axis('off')
    plt.show()

    return img_denoised


_img = '/content/photo_2025-02-09_11-33-20.jpg'

# Если изображение имеет небольшие шумы, попробуйте tau = 10, 20 и т.д.
denoise_image_with_svd_soft(_img, tau=10, use_color=True)
````
